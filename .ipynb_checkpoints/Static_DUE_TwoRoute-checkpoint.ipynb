# 1. Equilibrium Assignment on 2 Routes

```python
Q = 100
#f = 0.2 #in case of a fixed f
import math
import matplotlib.pyplot as plt

#Logit >> Stochastic equilibrium assignment!

#Function description:

FC_1 = 30
FC_2 = 10

VarC_1 = 3
VarC_2 = 9

# Empty lists for retrieval

lst_splits1_stoch = []
lst_splits2_stoch = []

lst_splits1_aon = []
lst_splits2_aon = []


def costs_given_flowsplit(split1=1,split2=0):
    # calculate the cost of the different routes,
    # given the splits

    flow_on_route1 = split1*Q
    flow_on_route2 = split2*Q

    cost_route_1 = FC_1 + VarC_1*flow_on_route1
    cost_route_2 = FC_2 + VarC_2*flow_on_route2

    return cost_route_1, cost_route_2

def flowsplit_given_costs(cost1=0,cost2=0):
    #given costs or disutilities, calculate the flowsplits

    FS_on_route1 = math.exp(-cost1) / (math.exp(-cost1) + math.exp(-cost2))
    FS_on_route2 = math.exp(-cost2) / (math.exp(-cost1) + math.exp(-cost2))

    return FS_on_route1, FS_on_route2

def flowsplit_aon_given_costs(cost1=0, cost2=0):

    min_of_costs = min(cost1,cost2)
    if min_of_costs == cost1:
        FS_on_route1 = 1
        FS_on_route2 = 0
    else:
        FS_on_route1 = 0
        FS_on_route2 = 1

    return FS_on_route1, FS_on_route2

def equilibrium_stoch(split1=1, split2=0, i=0):

    (cost1,cost2) = costs_given_flowsplit(split1, split2)
    (fs1, fs2) = flowsplit_given_costs(cost1,cost2)
    #The condition for convergence
    if i == 500:
        print((fs1*Q,fs2*Q))
    else:
        i += 1
        f = 1/i
        W_split1 = (1-f)*split1 + f*fs1
        W_split2 = (1-f)*split2 + f*fs2
        lst_splits1_stoch.append(W_split1)
        lst_splits2_stoch.append(W_split2)
        equilibrium_stoch(W_split1, W_split2, i)

def equilibrium_aon(split1=1, split2=0, j=0):

    (cost1,cost2) = costs_given_flowsplit(split1, split2)
    (fs1, fs2) = flowsplit_aon_given_costs(cost1,cost2)
    #The condition for convergence
    if j == 500:
        print((split1*Q,split2*Q))
    else:
        j += 1
        f = 1/j
        W_split1 = (1-f)*split1 + f*fs1
        W_split2 = (1-f)*split2 + f*fs2
        lst_splits1_aon.append(W_split1)
        lst_splits2_aon.append(W_split2)
        equilibrium_aon(W_split1, W_split2, j)

equilibrium_stoch(0.5, 0.5, 0)
equilibrium_aon(0.5, 0.5, 0)

plt.figure()

plt.subplot(121)
plt.plot(lst_splits1_stoch)
plt.plot(lst_splits2_stoch)
plt.ylabel('split on route 1 and 2 (Stoch)')
plt.xlabel('iteration')

plt.subplot(122)
plt.plot(lst_splits1_aon)
plt.plot(lst_splits2_aon)
plt.ylabel('split on route 1 and 2 (AoN)')
plt.xlabel('iteration')

plt.show()
```

# 2. Equilibrium Assignment on 3 Routes

```python
Q = 100
#f = 0.2 #in case of a fixed f
import math
import matplotlib.pyplot as plt

#Logit >> Stochastic equilibrium assignment!

#Function description:

#FILL COEFFICIENTS IN HERE
FC =   [27,    18,    27   ]
VarC = [0.05,  0.05,  0.05]

# Empty lists for retrieval

lst_splits1_stoch = []
lst_splits2_stoch = []
lst_splits3_stoch = []

lst_splits1_aon = []
lst_splits2_aon = []
lst_splits3_aon = []


def costs_given_flowsplit(split1=1, split2=0, split3=0):
    # calculate the cost of the different routes,
    # given the splits

    flow_on_route1 = split1*Q
    flow_on_route2 = split2*Q
    flow_on_route3 = split3*Q

    cost_route_1 = FC[0] + VarC[0]*(flow_on_route1 + flow_on_route2)
    cost_route_2 = FC[1] + VarC[1]*(flow_on_route1 + flow_on_route2) + VarC[2]*(flow_on_route2 + flow_on_route3)
    cost_route_3 = FC[2] + VarC[2]*(flow_on_route2 + flow_on_route3)

    return (cost_route_1, cost_route_2, cost_route_3)

def flowsplit_given_costs(cost1=0, cost2=0, cost3=0):
    #given costs or disutilities, calculate the flowsplits

    FS_on_route1 = math.exp(-cost1) / (math.exp(-cost1) + math.exp(-cost2) + math.exp(-cost3))
    FS_on_route2 = math.exp(-cost2) / (math.exp(-cost1) + math.exp(-cost2) + math.exp(-cost3))
    FS_on_route3 = math.exp(-cost3) / (math.exp(-cost1) + math.exp(-cost2) + math.exp(-cost3))

    return (FS_on_route1, FS_on_route2, FS_on_route3)

def flowsplit_aon_given_costs(cost1=0, cost2=0, cost3=0):

    min_of_costs = min(cost1,cost2,cost3)
    if min_of_costs == cost1:
        FS_on_route1 = 1
        FS_on_route2 = 0
        FS_on_route3 = 0
    elif min_of_costs == cost2:
        FS_on_route1 = 0
        FS_on_route2 = 1
        FS_on_route3 = 0
    else:
        FS_on_route1 = 0
        FS_on_route2 = 0
        FS_on_route3 = 1

    return (FS_on_route1, FS_on_route2, FS_on_route3)

def equilibrium_stoch(split1=1, split2=0, split3=0, i=0):

    (cost1,cost2,cost3) = costs_given_flowsplit(split1, split2, split3)
    (fs1, fs2, fs3) = flowsplit_given_costs(cost1, cost2, cost3)
    #The condition for convergence
    if i == 500:
        return print((fs1*Q, fs2*Q, fs3*Q))
    else:
        i += 1
        f = 1/i
        W_split1 = (1-f)*split1 + f*fs1
        W_split2 = (1-f)*split2 + f*fs2
        W_split3 = (1-f)*split3 + f*fs3

        lst_splits1_stoch.append(W_split1)
        lst_splits2_stoch.append(W_split2)
        lst_splits3_stoch.append(W_split3)

        equilibrium_stoch(W_split1, W_split2, split3, i)

def equilibrium_aon(split1=1, split2=0, split3=0, j=0):

    (cost1, cost2, cost3) = costs_given_flowsplit(split1, split2, split3)
    (fs1, fs2, fs3) = flowsplit_aon_given_costs(cost1, cost2, cost3)
    #The condition for convergence
    if j == 500:
        return print((split1*Q, split2*Q, split3*Q))
    else:
        j += 1
        f = 1/j

        W_split1 = (1-f)*split1 + f*fs1
        W_split2 = (1-f)*split2 + f*fs2
        W_split3 = (1-f)*split2 + f*fs3

        lst_splits1_aon.append(W_split1)
        lst_splits2_aon.append(W_split2)
        lst_splits3_aon.append(W_split3)

        equilibrium_aon(W_split1, W_split2, W_split3, j)

equilibrium_stoch(1, 0, 0, 0)
equilibrium_aon(1, 0, 0, 0)

plt.figure()

plt.subplot(121)
plt.plot(lst_splits1_stoch)
plt.plot(lst_splits2_stoch)
plt.plot(lst_splits3_stoch)
plt.ylabel('split on route 1 and 2 (Stoch)')
plt.xlabel('iteration')

plt.subplot(122)
plt.plot(lst_splits1_aon)
plt.plot(lst_splits2_aon)
plt.plot(lst_splits3_aon)
plt.ylabel('split on route 1 and 2 (AoN)')
plt.xlabel('iteration')

plt.show()
```
